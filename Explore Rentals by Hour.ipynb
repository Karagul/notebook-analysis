## Continuous Features (temp, atemp, hum, windspeed)

```python
cols = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
g = sb.pairplot(df_hour, vars=cols, hue='weathersit', markers='.', kind='reg')
for ax in g.axes[:4,:4].flat:
    ax.set_ylim((-.1, 1.1))
for ax in g.axes[4,:4].flat:
    ax.set_ylim((0, 1000))
```
