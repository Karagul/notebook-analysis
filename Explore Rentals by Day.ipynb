# Setup

```python
import pandas as pd 
import seaborn as sb
import numpy as np
%matplotlib inline
```

```python
# Load
df_day = pd.read_csv('data/raw/day.csv')
```

## Target Variable (cnt)

```python
sb.distplot(df_day.cnt)
```

## Continuous Features (temp, atemp, hum, windspeed)

```python
cols = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
g = sb.pairplot(df_day, vars=cols, hue='weathersit', markers='.', kind='reg')
for ax in g.axes[:4,:4].flat:
    ax.set_ylim((-.1, 1.1))
```

Rental count looks to be positively corelated with temp/atemp, and negatively correlated with windspeed

## Categorical Features (season, workingday, weathersit)

```python
g = sb.factorplot(x="season", y="cnt", hue="weathersit", col="workingday",
                   data=df_day, kind="swarm", size=6, aspect=1, split=True)
```

```python
g = sb.factorplot(x="season", y="cnt", hue="weathersit", col="workingday",
                   data=df_day, kind="violin", size=6, aspect=1)
```

# Findings

* Weather has an obvious impact on the rental count as well as season. 
* Rental count looks to be positively corelated with temp/atemp, and negatively correlated with windspeed
